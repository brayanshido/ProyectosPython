2.¿Cómo se distribuyen los títulos según su tipo (película, serie)?

```python
tipos=disney_data['type'].value_counts()
print("Distribución según tipo:")
print(tipos)
```

```python
tipo = disney_data['type'].value_counts()
print(tipo)
#  porcentajes
total = len(disney_data)
porcentaje_peliculas = (tipo['Movie'] / total) * 100
porcentaje_series = (tipo['TV Show'] / total) * 100
print(f'Porcentaje de películas: {porcentaje_peliculas:.1f}%')
print(f'Porcentaje de series: {porcentaje_series:.1f}%')
plt.show()

#Crear gráfico de barras
plt.figure(figsize=(10,30)),
sns.catplot(
    data=disney_data,
    x='type',
    kind='count'
)
plt.show()

# Crear un gráfico de pastel
plt.figure(figsize=(6, 6))
plt.pie(tipo, labels=tipo.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Títulos')
plt.axis('equal')
plt.show()
```

```python
# Crear un diccionario para almacenar la distribución de títulos por año y tipo
distribucion_por_tipo = {}

# Iterar a través de las filas del DataFrame
for index, row in disney_data.iterrows():
    titulo = row['title']
    tipo = row['type']

    if tipo in distribucion_por_tipo:
        distribucion_por_tipo[tipo].append(titulo)
    else:
        distribucion_por_tipo[tipo] = [titulo]

# Imprimir la distribución de títulos por año y tipo en orden ascendente
for tipo, titulos_en_tipo in distribucion_por_tipo.items():
    print(f"Tipo: {tipo}")
    for titulo in titulos_en_tipo:
        print(f"- {titulo}")
```

3.¿Cuál es la distribución de títulos a lo largo de los años?

```python
años=disney_data['release_year'].value_counts()
print("Distribución según años")
print(años)
```

```python
#Histograma de plt
plt.figure(figsize=(10,30)),
sns.catplot(
    data=disney_data,
    x='release_year',
    kind='count'
)
plt.show()

#Histograma que agrupa bien la data
plt.hist(disney_data['release_year'], color='red')
plt.show()
```

```python

#Histograma más ordenado
#Forma 1 de contar
año_titulo = disney_data["release_year"].value_counts()

#Forma 2 de contar
año_titulo= disney_data.groupby("release_year")['release_year'].count()

#1er método: Enseñado por docente
año = año_titulo.index
cantidad = año_titulo.values

plt.figure(figsize=(20,6)) #Alto y Ancho esta medido en pulgadas y esto se
plt.bar(año, cantidad,color=('red','blue','green'))
plt.show()
```

```python
#2do método: Aprendido por ahí y que es más profesional

año_titulo.plot(kind='bar', figsize=(20, 6))
plt.title('Distribución de titulos por año')
plt.xlabel('AÑO')
plt.ylabel('CANTIDAD DE PELICULAS')
plt.show()
```

```python
# Crear un diccionario para almacenar la distribución de títulos por año
distribucion_por_año = {}

# Iterar a través de las filas del DataFrame
for index, row in  disney_data.iterrows():
    año = row['release_year']
    titulo = row['title']

    if año in distribucion_por_año:
        distribucion_por_año[año].append(titulo)
    else:
        distribucion_por_año[año] = [titulo]

# Imprimir la distribución de títulos por año
for año, titulos_en_año in sorted(distribucion_por_año.items()):
    print(f"Año {año}:")
    for titulo in titulos_en_año:
        print(f"- {titulo}")
```

7.¿Cómo se distribuyen las calificaciones de los títulos?

```python
rating=disney_data['rating'].value_counts()
print("Distribución según rating:")
print(rating)

# Crear un gráfico de pastel
plt.figure(figsize=(6, 6))
plt.pie(rating, labels=rating.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de rating')
plt.show()

#Crear gráfico de barras
plt.figure(figsize=(10,30)),
sns.catplot(
    data=disney_data,
    x='rating',
    kind='count'
)
plt.show()
```
